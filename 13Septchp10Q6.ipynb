```python
#two columns for the two populations to be replaced by two csv files
population1= hotel1['Atlanta'] #replace with the actual column name
population2= hotel2['Houston'] #replace with the actual column name
```

```python
#calculate means and data sample size
mean1 = np.mean(population1)
mean2 = np.mean(population2)
n1=len(population1)
n2=len(population2)
sigma1=20
sigma2=25
print(n1,n2)
print(mean1,mean2)
```

```python
#specify the hypothesis
print("Ho: mu1-mu2>=0 v/s Ha: mu1-mu2<0")
```

```python
#calculate point estimate, standard error and z-statistics
point_estimate=mean1-mean2
print(point_estimate)
std_error=np.sqrt((sigma1**2/n1) + (sigma2**2/n2))
print(std_error)
zobs=point_estimate/std_error
zobs
```

```python
#Determine the critival value or p-value
alpha= 0.05 #significance level
zcri = norm.ppf(alpha)
print("zobs=:",zobs)
print("zcri=:",zcri)
```

```python
#calculate p-value
p_value = norm.cdf(-abs(zobs))
print("p_value=:", p_value)
```

```python
#make decision based on critical value
if abs(zobs) > abs(zcri):
    print(f"refect the null hypothesis (zobs={zobs:.2f}, p-value={p_value:.4f})")
else:
    print(f"Fail to reject the null hypothesis (zobs ={zobs:.2f}, p-value={p_value:.4f})")
```
