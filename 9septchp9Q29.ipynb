```python
# Question 29 page 443
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import t
```

```python
data = pd.read_csv("C:/Users/asus/Desktop/SMDA solved/PythonStatistics - SMDA/residentialwater.csv")
```

```python
column_name = "Rate (5 CCF)"
data_column = data[column_name]
```

```python
mu = 21.62
print("Ho: mu=21.62 v/s Ha: mu<>21.62")
```

```python
xbar = data_column.mean()
std = data_column.std()
n = len(data_column)
standard_error= std/np.sqrt(n)
```

```python
tobs = (xbar - mu) / standard_error
print("tobs:", tobs)
degrees_of_freedom = n - 1
significance_level = 0.05

tcriL = t.ppf(significance_level/2,degrees_of_freedom)
tcriU = t.ppf(1-significance_level/2,degrees_of_freedom)
print("tcriL:", tcriL)
print("tcriU:", tcriU)
```

```python
p_value = t.cdf(-abs(tobs), degrees_of_freedom)*2
print("p_value:", p_value)
```
