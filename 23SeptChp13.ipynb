```python
#Question 6 page 613 #One way Anova
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df=pd.read_csv("C:/Users/asus/Desktop/SMDA solved/PythonStatistics - SMDA/exer6.csv")
df
```

```python
#state the hypothesis
#Ho: All means are equal
#Ha: Atleast one mean is different
```

```python
mod=ols('values~treatment',data=df).fit()    #we have to find mean therefore written first(metric variable) and the categorical variable will come later (eg. values~treatment)
```

```python
#perform one-way ANOVA
aov_table=sm.stats.anova_lm(mod,type=2)   #type 2 is one way anova (we will alaways use this all ques) , Type 1 is two way anova 
print(aov_table)
```

```python
#Read the important values
Fobs=5.84486
pvalue=0.008269
print(f"F-statistic:{Fobs}")
print(f"pvalue: {pvalue}")
```

```python
#Interpret the p-value
alpha = 0.05
if pvalue<alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
    print("Perform Fisher LSD to check which group is different")
else:
    print("Fail to  reject the null hypothesis: There is no signifocant difference between the groups.")
```

```python
#If Anova shows significant results, proceed with post-hoc test
pair_t=mod.t_test_pairwise('treatment')     
pair_t.result_frame
```

```python
#as p<alpha for difference between treatments A&C   (since true, means averages are different) 
#There is signnificant difference between the emans for treatments A&C
```

```python
#Question 10 page 614 #One way Anova
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df=pd.read_csv("C:/Users/asus/Desktop/SMDA solved/PythonStatistics - SMDA/audjudg.csv")
df
```

```python
#state the hypothesis
#Ho: All means are equal
#Ha: Atleast one mean is different
```

```python
mod=ols('errors~method',data=df).fit()  
```

```python
#perform one-way ANOVA
aov_table=sm.stats.anova_lm(mod,type=2)  
print(aov_table)
```

```python
#Read the important values
Fobs=5.84486
pvalue=0.008269
print(f"F-statistic:{Fobs}")
print(f"pvalue: {pvalue}")
```

```python
#Interpret the p-value
alpha = 0.05
if pvalue<alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
    print("Perform Fisher LSD to check which group is different")
else:
    print("Fail to  reject the null hypothesis: There is no signifocant difference between the groups.")
```

```python
#If Anova shows significant results, proceed with post-hoc test
pair_t=mod.t_test_pairwise('method')     
pair_t.result_frame
```

```python
#as p<alpha for  all methods  (since true, means averages are different) 
#There is signnificant difference between the means for all methods
```
